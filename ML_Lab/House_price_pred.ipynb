```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('E:\Machine-Learning\Dataset\Mumbai House Prices.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
# drop unnecessary column from dataframe

# df = df.drop('Status',axis=1)
# df = df.drop('Landmarks',axis=1)

#Check that now data is clean or not 

# data has null value of balcony so assume that that house has no balcony so feel it with 0 
df['Balcony'].replace([np.nan],0,inplace=True)

# data has null value of parking so assume that that house has no facality of parking it with0 
df['parking'].replace([np.nan],0,inplace=True)

# data has null value of parking so assume that that house has no parking so feel it with 0 
df['Lift'].replace([np.nan],0,inplace=True)

df.head()

df.isnull().sum()

# df.head()
```

```python
# label encoder is only apply on object datatypes

df['Furnished_status'].unique()
df['Furnished_status'].value_counts()
```

```python
label_encoder = LabelEncoder()

# using label encoder we've changed the strig to int of furnished status
df['Furnished_status'] = label_encoder.fit_transform(df['Furnished_status'])

# now fill that value with unfurnished

df.head()
```
