```python
import tensorflow
from tensorflow import keras
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_valid_data_generator = ImageDataGenerator(rescale=1./255,
                                                validation_split=.2)
train_set = train_valid_data_generator.flow_from_directory('E:/Machine-Learning/ML Club/sports-balls-multiclass-image-classification/train/',
target_size=(256,256),
color_mode="rgb",
class_mode='categorical',
batch_size=32,
subset="training"
)

valid_set = train_valid_data_generator.flow_from_directory("E:/Machine-Learning/ML Club/sports-balls-multiclass-image-classification/train/",
target_size=(256,256),
color_mode="rgb",
class_mode='categorical',
batch_size=32,
subset='validation'
)

test_data_generator = ImageDataGenerator()

test_set = test_data_generator.flow_from_directory('E:/Machine-Learning/ML Club/sports-balls-multiclass-image-classification/test/',
target_size=(256,256),
color_mode='rgb',
class_mode='categorical',
batch_size=32)
```
